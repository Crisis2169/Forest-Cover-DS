```python
import os        #helps interact with your computer files/folders.
import pandas as pd      #loading and working with datasets
import numpy as np    #for mathematical operations

# plotting
import matplotlib.pyplot as plt #make graphs
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import seaborn as sns

# preprocessing & modeling
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import joblib                     #used to save trained model
```

```python
DATA_PATH = r"C:\Users\jiten\Downloads\covtype.csv"
df = pd.read_csv(DATA_PATH)
print("Loaded dataset shape:", df.shape)
display(df.head())
```

```python
# 1. Detect target column
if "Cover_Type" in df.columns:
    target_col = "Cover_Type"
else:
    target_col = df.columns[-1]

print("Target column:", target_col)
```

```python
# 2. Define X and y_orig
X = df.drop(columns=[target_col])
y_orig = df[target_col]
```

```python
# 3. Label Encode target
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y_orig)
```

```python
# 4. Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
preprocessor = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),        #Fills missing values in numeric columns.
    ("scaler", StandardScaler())     #rescale numbers to have comparable range
])

X_train_pre = preprocessor.fit_transform(X_train[numeric_cols])
X_test_pre = preprocessor.transform(X_test[numeric_cols])
```

```python
rf = RandomForestClassifier(
    n_estimators=200,          #model will build 200 decision trees.
    max_depth=20,             #Each tree can grow up to depth 20
    n_jobs=-1,
    random_state=42
)
rf.fit(X_train_pre, y_train)
y_pred_rf = rf.predict(X_test_pre)
rf_acc = accuracy_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf, average='macro')

print(f"RandomForest  -> accuracy: {rf_acc:.4f}, macro F1: {rf_f1:.4f}")
```

```python
# 6. XGBoost Model
import xgboost as xgb
from sklearn.metrics import f1_score, accuracy_score

#Count number of classes
n_classes = len(np.unique(y_train))

xgb_clf = xgb.XGBClassifier(
 
    n_estimators=300,        #Build 300 boosted trees
    max_depth=8,             #Each tree has max depth 8
    learning_rate=0.05,
    n_jobs=-1,
    random_state=42
)

# First fit the model
xgb_clf.fit(X_train_pre, y_train)

# Then make predictions
y_pred = xgb_clf.predict(X_test_pre)  # Make sure X_test_pre is defined

# Now calculate metrics using the predictions
xgb_acc = accuracy_score(y_test, y_pred)
xgb_f1 = f1_score(y_test, y_pred, average='macro')

print(f"XGBoost      -> accuracy: {xgb_acc:.4f}, macro F1: {xgb_f1:.4f}")
```

```python
xgb_acc = xgb_clf.score(X_test, y_test)  # Assuming xgb_clf is already defined
rf_acc = rf.score(X_test, y_test)        # Assuming rf is already defined

# Now use the defined variables in the comparison
if xgb_acc >= rf_acc:
    best_model = xgb_clf
    best_name = "xgboost_covtype.pkl"
else:
    best_model = rf
    best_name = "rf_covtype.pkl"

joblib.dump({
    "model": best_model,
    "preprocessor": preprocessor,
    "numeric_cols": numeric_cols,
    "label_encoder": le
}, best_name)

print(f"Saved best model and artifacts to: {best_name}")
```

```python
y_test_orig = le.inverse_transform(y_test)     #Convert encoded y_test back to original labels
y_pred_best = best_model.predict(X_test_pre)      #Predict using the best model
y_pred_orig = le.inverse_transform(y_pred_best)  #Convert predicted encoded labels back to original values

print("\nClassification report (with original label values):")
print(classification_report(y_test_orig, y_pred_orig, digits=4))
```

```python
sample = X_test_pre[:5]      #Select the first 5 test samples
print("Sample predictions:", best_model.predict(sample))

decoded_preds = le.inverse_transform(best_model.predict(sample))
print("Original label predictions:", decoded_preds)
```
